```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Medium', 'High']
df['Income Band'] = pd.cut(df['Estimated Income'],bins=bins,labels=labels)
```

```python
df['Income Band'].value_counts().plot(kind="bar")
```

```python
df.columns
```

```python
#Examine the distribution of unique categories in categorical columns
categorical_cols=["Nationality", "Occupation", "Loyalty Classification", "Amount of Credit Cards", "Properties Owned", "Risk Weighting", "BRId", "GenderId", "IAId", "Income Band"]
```

```python

for col in categorical_cols:
  print(f"Value counts for {col} :")
  display(df[col].value_counts())
```

```python
#Univariant Analysis
for i, predictor in enumerate(categorical_cols):
  plt.figure(i)
  sns.countplot(data=df, x=predictor)
```

```python
#Bivariant Analysis
for i, predictor in enumerate(categorical_cols):
  plt.figure(i)
  sns.countplot(data=df, x=predictor, hue="Nationality")
```

```python
#Histplot of value counts for different occupations
for col in categorical_cols:
  if col=="Occupation":
    continue
  plt.figure(figsize=(8,4))
  sns.histplot(df[col])
  plt.title('Histogram of Occupation Count')
  plt.xlabel(col)
  plt.ylabel("Count")
  plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df["Estimated Income"])
plt.title('Histogram of Estimated Income')
plt.xlabel('Estimated Income')
plt.ylabel("Count")
plt.show()
```

HEATMAPS

```python
numerical_cols= ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts','Saving Accounts', 'Foreign Currency Account', 'Business Lending']

correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

##INSIGHTS OF EDA:

1. The strongest positive correlation is between Bank Deposits and Checking Accounts (0.84), indicating that customers with higher deposits also maintain higher checking account balances.

2. Bank Deposits also show a strong positive correlation with Saving Accounts (0.75), suggesting that customers with larger deposits tend to maintain substantial savings.

3. Estimated Income has a moderate positive correlation with most banking products, indicating that higher-income customers generally utilize more financial services.

4. No negative correlations are observed, implying that the banking products complement each other rather than having inverse relationships.
